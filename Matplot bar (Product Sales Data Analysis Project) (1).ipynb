```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
Product= ["CPU", "Monitor" , "Keyboard" , "Mouse"]
Jan= [50,100,300,200]
plt.bar(Product,Jan)
```

```python
Product= ["CPU", "Monitor" , "Keyboard" , "Mouse"]
Jan= [50,100,300,200]
plt.bar(Product,Jan)
plt.xlabel("Product",fontsize=15)
plt.ylabel(" product in Jan",fontsize=15)
plt.title("sale of computer parts",fontsize=15)
```

```python
Product= ["CPU", "Monitor" , "Keyboard" , "Mouse"]
Jan= [50,100,300,200]
plt.bar(Product,Jan,color="g",edgecolor="r",linewidth=3,)
plt.xlabel("Product",fontsize=15)
plt.ylabel(" product in Jan",fontsize=15)
plt.title("sale of computer parts",fontsize=15)
```

```python
Product= ["CPU", "Monitor" , "Keyboard" , "Mouse"]
Jan= [50,100,300,200]
plt.bar(Product,Jan,color="g",edgecolor="r",linewidth=3,linestyle="--")
plt.xlabel("Product",fontsize=15)
plt.ylabel(" product in Jan",fontsize=15)
plt.title("sale of computer parts",fontsize=15)
```

```python
Product= ["CPU", "Monitor" , "Keyboard" , "Mouse"]
Jan= [50,100,300,200]
Feb= [120,220,400,300]
plt.bar(Product,Jan,color="g",edgecolor="r",linewidth=3,label="Jan")
plt.xlabel("Product",fontsize=15)
plt.ylabel(" product in Jan",fontsize=15)
plt.title("sale of computer parts",fontsize=15)
plt.legend()
```

```python
Product= ["CPU", "Monitor" , "Keyboard" , "Mouse"]
Jan= [50,100,300,200]
Feb= [120,50,100,300]
width=0.2
Product_values=np.arange(len(Product))
plt.bar(Product_values,Jan,width=0.2,color="g",label="Jan")
plt.bar(Product_values+width,Feb,width=0.2,color="r",label="Feb")
plt.xticks(Product_values+width,Product)
plt.xlabel("Product",fontsize=15)
plt.ylabel(" product in Jan",fontsize=15)
plt.title("sale of computer parts",fontsize=15)
plt.legend()
```

```python
Product= ["CPU", "Monitor" , "Keyboard" , "Mouse"]
Jan= [50,100,300,200]
Feb= [120,50,100,300]
Mar= [60,150,320,280]
width=0.2
Product_values=np.arange(len(Product))
plt.bar(Product_values,Jan,width=0.2,color="g",label="Jan")

plt.bar(Product_values+width,Feb,width=0.2,color="r",label="Feb")

plt.bar(Product_values+width+width,Mar,width=0.2,color="b",label="Mar")

plt.xticks(Product_values+width,Product)
plt.xlabel("Product",fontsize=15)
plt.ylabel(" product in Jan",fontsize=15)
plt.title("sale of computer parts",fontsize=15)
plt.legend()
```

```python
Product= ["CPU", "Monitor" , "Keyboard" , "Mouse"]
Jan= [50,100,300,200]
Feb= [120,50,100,300]
Mar= [60,150,320,280]
width=0.2
Product_values=np.arange(len(Product))
plt.barh(Product_values,Jan,width,color="g",label="Jan")

plt.barh(Product_values+width,Feb,width,color="r",label="Feb")

plt.barh(Product_values+width+width,Mar,width,color="b",label="Mar")

plt.yticks(Product_values+width,Product)
plt.xlabel("Product",fontsize=15)
plt.ylabel(" product in Jan",fontsize=15)
plt.title("sale of computer parts",fontsize=15)
plt.legend()
```
